# What it means

<font size="4"> 

The magic happens in the return_pz_stats file. We define the interpolate statistics with variables of user limited magnitudes
and reshifts. This task will return the redshift bin that this user-supplied redshift falls 
into and a 5-element array of interpolated values.
Then, Figure out which redshift bin (zbins) the user-supplied redshift falls into. Next initialize the 5 element array of 
interpolated statistics that the task will return. 
Lastly, for each of the statistics, load the pickle file for this zbin and interpolate.

Here are the accessible ranges for limiting magnitudes and red shift:


    filters = ['u','g','r','i','z','y']
    magmins = [23.15, 24.25, 23.95, 23.25, 22.55, 21.35]
    magmaxs = [26.78, 27.88, 27.84, 27.15, 26.37, 25.17]
    zbinmin = 0.30
    zbinmax = 3.00
    
This means only values in these ranges will be accepted. If the values are out of this range you will get
an error message. 
The redshift returned is a bin which the input redshift falls into.
Depedning on the input, you can easily change the return value you want given. 

It is returned as a 5 element array of standard deviation, its error, bias, its error, and fraction of outliers
[stddevs,stddevserr,biases, biaseserr,fouts].



# Example 1
<font size="4"> For our first example, lets calculate some values. First start off with an array of limiting magnitudes.
Each bin should be a baseline for ugrizy filters. We then insert the reshift we want to use. In this example we use 0.3. 
We want to find the results for all 5 bins. To do this, just change the number in the results_stats[]. 
Last we delete what we dont need.


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import return_pz_stats
import warnings
warnings.filterwarnings('ignore')
```

```python
# Insert the limiting magnitudes here
mags = np.asarray( [26.085, 27.379, 27.531, 26.831, 26.055, 24.855], dtype='float' )
 
results_zbin,results_stats = return_pz_stats.interpolate_statistics( mags, 0.3 ) # This is where you input redshift value. 
stddev = results_stats[0]    # Changing this number (0-4) here will give you the different results discussed above.
stddev_error = results_stats[1]
bias = results_stats[2]
bias_error = results_stats[3]
fractionofoutliers = results_stats[4]

# Results here
print('Predicted statistical measures (value, error):')
print('standard deviation   %10.6f %10.6f ' % (results_stats[0],results_stats[1]))
print('bias                 %10.6f %10.6f ' % (results_stats[2],results_stats[3]))
print('fraction of outliers %10.6f ' % (results_stats[4]))

del results_zbin,results_stats
```

# Example 2

# Example #3

<font size="4">
The example shows g-band filter at a redshift of 0.3, using a baseline set of
limiting magnitudes that represent the LSST depth after 10 years 
for a survey that spends half/double the amount of time on g-band exposures
We will follow the same guidelines as above (with some small renaming) to find the results for all the bins.</font>

```python

mags = np.asarray( [26.085, 27.379, 27.531, 26.831, 26.055, 24.855], dtype='float' )


gmag_hlf = 27.003
gmag_dbl = 27.755

#Change the # in "mags[]" to 1 because we are replacing the magnitude in the "mags" array with half/dbl g_mag numbers
mags[1] = gmag_hlf 
results_zbin,results_stats = return_pz_stats.interpolate_statistics( mags, 0.3 )
stddev_half = results_stats[0]
stddev_error_half = results_stats[1]
bias_half = results_stats[2]
bias_error_half = results_stats[3]
fractionofoutliers_half = results_stats[4]

print('Predicted statistical measures half time (value, error):')
print('standard deviation   %10.6f %10.6f ' % (results_stats[0],results_stats[1]))
print('bias                 %10.6f %10.6f ' % (results_stats[2],results_stats[3]))
print('fraction of outliers %10.6f ' % (results_stats[4]))

del results_zbin,results_stats

mags[1] = gmag_dbl
results_zbin,results_stats = return_pz_stats.interpolate_statistics( mags, 0.3 )
stddev_dbl = results_stats[0]
stddev_error_dbl = results_stats[1]
bias_dbl = results_stats[2]
bias_error_dbl = results_stats[3]
fractionofoutliers_dbl = results_stats[4]

print('Predicted statistical measures double time (value, error):')
print('standard deviation   %10.6f %10.6f ' % (results_stats[0],results_stats[1]))
print('bias                 %10.6f %10.6f ' % (results_stats[2],results_stats[3]))
print('fraction of outliers %10.6f ' % (results_stats[4]))

del results_zbin,results_stats
```

```python
# Plotting:
<font size="4"> Now lets show the results of standard deviation with its error, bias with its error, and fraction of outliers
all in plot form. First we need to set up our arrays that hold the values we need. 
Next, we want to calculate the times spent in g-band filter. We do this by defining r as our red shift bin.
Then we just repeat steps we did in the previous examples.
When setting up the plots, we want to use plot.errorbar so the error results show up nicely. For this use plot.error.
</font>
```

```python
# Make arrays to hold the values we need
half_stddev = np.zeros( 27, dtype='float' )
dbl_stddev = np.zeros( 27, dtype='float' )
half_stderror = np.zeros( 27, dtype='float' )
dbl_stderror = np.zeros( 27, dtype='float' )
half_bias = np.zeros( 27, dtype='float' )
dbl_bias = np.zeros( 27, dtype='float' )
half_biaserror = np.zeros( 27, dtype='float' )
dbl_biaserror = np.zeros( 27, dtype='float' )
half_fouts = np.zeros( 27, dtype='float' )
dbl_fouts = np.zeros( 27, dtype='float' )

redshifts = np.arange( 27, dtype='float' )*0.1 + 0.35
gmags_hlf = np.asarray( [26.085, 27.003, 27.531, 26.831, 26.055, 24.855], dtype='float' )
gmags_dbl = np.asarray( [26.085, 27.755, 27.531, 26.831, 26.055, 24.855], dtype='float' )


# Calculation for spedning half the time in g-band filter.
for r,redshift in enumerate(redshifts):
    results_zbin,results_stats = return_pz_stats.interpolate_statistics( gmags_hlf, redshift )
    half_stddev[r] = results_stats[0]
    half_stderror[r] = results_stats[1]
    half_bias[r] = results_stats[2]
    half_biaserror[r] = results_stats[3]
    half_fouts[r] = results_stats[4]
    
    del results_zbin,results_stats
    
#  Calculation for spedning double the amount of time in g-band filter. 
for r,redshift in enumerate(redshifts):
    results_zbin,results_stats = return_pz_stats.interpolate_statistics( gmags_hlf, redshift )
    dbl_stddev[r] = results_stats[0]
    dbl_stderror[r] = results_stats[1]
    dbl_bias[r] = results_stats[2]
    dbl_biaserror[r] = results_stats[3]
    dbl_fouts[r] = results_stats[4]
    
    del results_zbin,results_stats
    

# Plots of half/double redshifts, standard deviation w/error, Biases w/ error and Fraction of Outliers

plt.figure(figsize=(16,8))
plt.rcParams.update({'font.size':20})
plt.plot(redshifts, half_stddev, 'o', ms=10, alpha=0.5, mew=0, color='green', label='Half_Stddev')
plt.plot(redshifts, dbl_stddev, 'o', ms=10, alpha=0.5, mew=0, color='forestgreen', label='Double_stddev')
plt.errorbar(redshifts,half_stddev, yerr=half_stderror, marker = 'x', ms=10, alpha=0.5, mew=0, color='red', label='Half_Stddeverror') 
plt.xlabel('Photometric Redshift')
plt.ylabel('Results')
plt.title('Observing Standard Deviation in Filter G on Photo-z at LSST Year 10')
plt.legend(loc='upper left',numpoints=1,markerscale=2)
plt.show()

plt.figure(figsize=(16,8))
plt.rcParams.update({'font.size':20})
plt.plot(redshifts, half_bias, 'o', ms=10, alpha=0.5, mew=0, color='blue', label='Half_bias')
plt.plot(redshifts, dbl_bias, 'o', ms=10, alpha=0.5, mew=0, color='blue', label='Double_bias')
plt.errorbar(redshifts, half_bias, yerr=half_biaserror,marker = 'x', ms=10, alpha=0.5, mew=0, color='red', label='Half_biaserror')
plt.xlabel('Photometric Redshift')
plt.ylabel('Results')
plt.title('Observing Bias in Filter G on Photo-z at LSST Year 10')
plt.legend(loc='upper left',numpoints=1,markerscale=2)
plt.show()
plt.figure(figsize=(16,8))
plt.rcParams.update({'font.size':20})
plt.plot(redshifts, half_fouts, 'o', ms=10, alpha=0.5, mew=0, color='violet', label='Half_fouts')
plt.plot(redshifts, dbl_fouts, 'o', ms=10, alpha=0.5, mew=0, color='violet', label='Double_fouts')

plt.xlabel('Photometric Redshift')
plt.ylabel('Results')
plt.title('Observing Fraction of Outliers in Filter G on Photo-z at LSST Year 10')
plt.legend(loc='upper left',numpoints=1,markerscale=2)
plt.show()
```

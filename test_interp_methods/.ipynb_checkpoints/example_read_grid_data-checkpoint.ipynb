This notebook is a demonstration of how to read the grid of standard deviation results.

```python
### Import needed packages
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<br><br><br><br>
Read data from the results file into numpy arrays and print array information.

```python
### Read the photo-z results into numpy arrays
### The header of the file grid_stddev.dat explains the data in that file

### First read the limiting magnitude depths from grid_stddevs.dat for all grid points
###   read all rows of the first six columns
maglims = np.loadtxt( 'grid_stddevs.dat', dtype='float', usecols=(0,1,2,3,4,5))
np.info(maglims)
print('shape = ',maglims.shape)
print('number of rows = ',maglims.shape[0])
print('number of columns = ',maglims.shape[1])

print(' ')

### Read the standard deviation in lowest redshift bin from grid_stddevs.dat
###    read all rows of the 7th column
values = np.loadtxt( 'grid_stddevs.dat', dtype='float', usecols={6})
np.info(values)
print('shape = ',values.shape)
```

<br><br><br><br>
Visualize the data. <br>
The first row of the data file represents the photometric depth, and thus the redshift results, for a "baseline" LSST observing strategy. Every other row represents a deviation from that strategy in one or more filters. A change in the photometric depth of any filter will change the standard deviation of the redshift results. Let's plot how the standard deviation changes as the photometric depth is varied.

```python
### Initiate the figure
plt.figure(figsize=(16,8))
plt.rcParams.update({'font.size':20})

### Define a new array that is the difference in r-band limiting magnitude for each run (each row)
diff_rband = np.zeros( maglims.shape[0], dtype='float' )
for i in range(maglims.shape[0]):
    diff_rband[i] += maglims[i][2] - maglims[0][2]

### What does diff_rband really mean though?
### Let's consider by asking what it means when when diff_rband < 0:
###   maglims[i] < maglims[0]
###   the magnitude limit of run "i" is a lower value than that of the baseline run "i=0"
###   the photometric depth of the survey simulated in run "i" is brighter (=shallower, =worse)
### From the plot we can see that when diff_rband < 0, the standard deviation is larger (i.e., worse results)

### Plot the value of the standard deviation as a function of the difference in the r-band magnitude limit
plt.plot(diff_rband, values, 'o', ms=7, alpha=0.5, mew=0, color='red', label='values')
plt.plot(diff_rband[0], values[0], 'o', ms=15, alpha=1.0, mew=0, color='black', label='baseline')

### Label the axes, add a legend, set the axes limits.
plt.xlabel('Difference in r-band Limit Compared to Baseline Survey Magnitude Limits')
plt.ylabel('Standard Deviation')
plt.legend(loc='best',numpoints=1)
# plt.xlim([-0.1,1.4])
# plt.ylim([0.01,0.060])

### Show the plot and/or make a PNG file of the plot
plt.show()
#fig.savefig('example_read_grid_data_plot1.png')
```

<br><br><br>
The overall photo-z results depend on more than just the r-band depth. <br>
Let's instead plot the results as a function of the distance in the 6-dimensional space of limiting magnitudes.<br>
Note that now, our x-axis will always be a positive value. 

```python
### Initiate the figure
plt.figure(figsize=(16,8))
plt.rcParams.update({'font.size':20})

dist6d = np.zeros( maglims.shape[0], dtype='float' )
for i in range(maglims.shape[0]):
    for j in range(maglims.shape[1]):
        dist6d[i] += (maglims[i][j] - maglims[0][j])**2

plt.plot(np.sqrt(dist6d), values, 'o', ms=7, alpha=0.5, mew=0, color='brown', label='values')
plt.plot(0.0, values[0], 'o', ms=15, alpha=1.0, mew=0, color='black', label='baseline')

### Label the axes, add a legend, set the axes limits.
plt.xlabel('6D "Distance" from Baseline Survey Magnitude Limits')
plt.ylabel('Standard Deviation')
plt.legend(loc='best',numpoints=1)
plt.xlim([-0.1,8])
# plt.ylim([0.01,0.060])

### Show the plot and/or make a PNG file of the plot
plt.show()
#fig.savefig('example_read_grid_data_plot1.png')
```
